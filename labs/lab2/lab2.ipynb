### 5

Заметьте, что столбец `DebtRatio` содержит неправдоподобные данные. Только значения, соответствующие известному месячному доходу, являются отношениями. Остальные – абсолютные значения месячных выплат процентов. 

Исправьте данные, сделав все значения столбца `DebtRatio` абсолютными (умножьте их на `MonthlyIncome`).  Чтобы ваша программа быстро работала на полных данных, постарайтесь не использовать цикл.

#### *Функции, которые могут пригодиться при решении:*

Обращение к элементам DataFrame:
  * элемент: `data.loc[i, 'названиеСтолбца']`
  * столбец: `data['названиеСтолбца']`
  * подматрица: `data.loc[a:b, списокНазванийСтолбцов]`

Условная индексация:
* `data.loc[data['столбец'] > 20, списокНазванийСтолбцов]`

лучше писать так:

* `i = data['столбец'] > 20`  # вектор `True` и `False`
* `data.loc[i, 'названиеСтолбца']`

У подматриц номера строк наследуются от исходной.

* `pandas.isnull(скаляр или массив)` - проверка, является ли значение неопределенным (`NaN`)
* `pandas.notnull(скаляр или массив)` - проверка, является ли значение определенным (не `NaN`)

````python
# место для кода

# Для выполнения задания нам нужно исправить только те строки, 
# где `DebtRatio` — абсолютные выплаты, то есть, их нужно заменить на отношение, умножив их на `MonthlyIncome`.


# Создаем маску для строк, где DebtRatio > 1 (т.е. вероятно, абсолютные выплаты)
mask = data['DebtRatio'] > 1

# Исправляем эти строки
# Новое значение DebtRatio = DebtRatio * MonthlyIncome
data.loc[mask, 'DebtRatio'] = data.loc[mask, 'DebtRatio'] * data.loc[mask, 'MonthlyIncome']


print(data.head())
````

### 7

Вычислите средний ежемесячный доход и присвойте всем клиентам с неизвестным доходом полученное число.

*Функции, которые могут пригодиться при решении: `.mean()`*

*Другие описательные статистики:* https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats

````python
# место для кода
# считаем средний ежемесячный доход по всем клиентам, игнорируя пропущенные значения
average_income = data['MonthlyIncome'].mean()
````

````python
print(average_income)
````

````python

# заполняем пропуски в столбце `MonthlyIncome` полученным средним значением
data['MonthlyIncome'].fillna(average_income, inplace=True)
````

````python
# прошлая программа выдаёт ошибку потому, что мы используем chained assignment с `inplace=True`, что в будущем pandas будет работать иначе.
# Чтобы избавиться от предупреждения и сделать код более устойчивым, 
# будем присвоить результат обратно в DataFrame, а не использовать `inplace=True`. 
````

````python
data['MonthlyIncome'] = data['MonthlyIncome'].fillna(average_income) 
# Это выполнит ту же операцию, но без chained assignment и предупреждений.
````

Используя метод `groupby`, оцените вероятности невозврата кредита (`SeriousDlqin2yrs=1`) для различных значений количества иждивенцев (`NumberOfDependents`).  

Проделайте аналогичную процедуру для различных значений столбца `NumberRealEstateLoansOrLines`

*Подсказка:*
`data['столбец1'].groupby(data['столбец2']).mean()`  *-- расчет средних значений столбца1 по группам из столбца2*

````python
# место для кода


import matplotlib.pyplot as plt # нужна для графиков 
import seaborn as sns #  библиотека, построенная на основе Matplotlib, которая предоставляет высокоуровневый интерфейс для создания статистических графиков.
````

````python
# cоздадим столбец с бинарным признаком невозврата кредита
# Предполагается, что SeriousDlqin2yrs == 1 означает невозврат
# Воспользуемся методом groupby и mean, чтобы оценить вероятность
dependents_prob = data['SeriousDlqin2yrs'].groupby(data['NumberOfDependents']).mean()
````
